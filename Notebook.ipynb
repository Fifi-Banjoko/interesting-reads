#### 1.2.3 Pre-trained Transformer Models

Pre-trained Transformer models are advanced AI models designed to understand and generate human language. They use a specialized neural network architecture known as the Transformer, which excels at processing sequential data like text. These models undergo a two-phase training process.

In the pre-training phase, they are exposed to vast amounts of text data from the internet and other sources. During this phase, the model learns to predict the next word in a sentence or fill in gaps in text, enabling it to capture grammar, syntax, semantics, and even world knowledge. After pre-training, the model can be fine-tuned for specific natural language processing (NLP) tasks, like text classification, translation, or question-answering. Fine-tuning customizes the model's knowledge for particular applications with relatively small amounts of task-specific data.

Pre-trained Transformer models have gained immense popularity due to their versatility and remarkable performance in various NLP tasks. Some well-known models include BERT, GPT, RoBERTa, XLNet, and T5. They are open-source and widely used across industries. However, challenges like large model sizes, high computational demands, and ethical concerns related to biases in training data continue to be areas of active research and development in the field. In essence, pre-trained Transformer models have revolutionized NLP by providing a foundation for advanced language understanding and generation in AI systems.

#### 1.2.4 Ethical Concerns

Utilising AI, particularly models like GPT-2, to offer mental health support raises significant ethical concerns that require careful attention.

Firstly, privacy and confidentiality are paramount. AI chatbots collect and handle user data during interactions, necessitating robust data security and compliance with data protection regulations to safeguard sensitive information. Informed consent is equally critical; users must be fully informed about the AI's capabilities and limitations, its data usage policies, and the absence of human expertise, and they should willingly consent to engage with it.

Bias and fairness are crucial ethical issues. Pre-trained models can inherit biases from their training data, potentially leading to unfair or harmful responses in a mental health context. Bias mitigation strategies and continuous auditing to improve fairness are essential components of ethical AI design.

Accountability poses another challenge. Establishing clear guidelines for determining responsibility in case of AI errors or harm is necessary. Additionally, user vulnerability must be acknowledged, especially as individuals seeking mental health support may be in emotionally fragile states. The AI should be designed with empathy and sensitivity, avoiding content that could exacerbate emotional distress.

Human oversight is imperative. AI chatbots should include mechanisms for trained mental health professionals to intervene when needed, ensuring user safety. It is crucial to convey that AI chatbots are complementary to, rather than a replacement for, human mental health professionals.

Transparency and explainability are vital for building user trust. Users should understand the AI's responses and decision-making processes. Lastly, continual monitoring and improvement of the AI's performance, content, and responses are essential to maintain a safe and effective tool for mental health support.

In essence, the ethical concerns surrounding AI-driven mental health support necessitate a comprehensive approach that prioritizes user privacy, fairness, accountability, transparency, and well-being. Collaboration with mental health professionals and adherence to ethical guidelines are essential for responsible and effective AI-based mental health support.

### 2.2. Methodology

#### 2.2.1 Data Collection

he project focused on fine-tuning a language model using conversational datasets and social media data, with an emphasis on mental health-related content. Several publicly available datasets, including the HOPE, MEMO, DAIC-WoZ, ALONE, Counsel-Chat, and PAIR datasets, were considered for fine-tuning. Among these, the Counsel-Chat dataset emerged as the most suitable choice for its alignment with the project's objectives. Counsel-Chat provides a platform where therapists respond to client questions, and users can like helpful responses, offering a diverse dataset of responses to various mental health queries.

In addition to conversational data, the project incorporated social media content from Twitter and Reddit. These platforms were chosen due to their popularity for discussing mental health topics and sharing opinions and insights. To collect relevant data, the project employed web scraping techniques. Twitter data was gathered using the Scrapper API, targeting queries such as 'mental health,' 'depression,' 'anxiety,' and 'suicide.' Similarly, mental health-related subreddit post titles were scraped using the PRAW API for Reddit. The collected data from both platforms was stored in an SQL database, with separate dataframes created for each platform, forming the basis for subsequent data processing and analysis.

#### 2.2.2 Data Preprocessing

Following data collection, a comprehensive data preprocessing phase was initiated to prepare the dataset for fine-tuning the GPT-2 model. The initial focus was on retaining only important text-based columns, removing superfluous features, and eliminating missing values and duplicate entries to enhance dataset quality. Additionally, posts marked as '[Removed by Reddit]' were filtered out due to their lack of meaningful content.

Furthermore, while maintaining punctuation relevant for context and clarity, non-essential symbols like '#,' '@,' '<,' '>', and '/' were removed. Notably, stemming and lemmatization were omitted from the preprocessing pipeline, as GPT-2's extensive pre-training already equipped it with robust word-context comprehension. Implementing these linguistic simplifications risked disrupting the model's understanding and introducing out-of-vocabulary (OOV) words. Finally, tokenization, executed by the Hugging Face GPT2 tokenizer, was performed to prepare the data for fine-tuning, ensuring its suitability and relevance for subsequent model training.

#### 2.2.3 Model Fine-Tuning

The GPT2 small model was chosen as the pretrained model because it offers advantages such as lower resource requirements, faster inference times, cost-efficiency, sufficient language understanding, reduced overfitting, and scalability. These benefits make it suitable for projects with hardware constraints such as this one, enabling real-time interactions, cost-effective training and deployment, while still providing meaningful responses for various text generation tasks

Before fine-tuning the model, datasets were created for both corpuses of text. The datasets were loaded using Pytorch's dataloader function that is used to load batches of data from a dataset during training. The first dataset containing dialogue data was first used to fine-tune the model since it is the most relevant data. Hyper-parameter tuning was performed to find the best hyper-parameters for the model before model training. This returned the training loss and validation loss at each step. After this, the model was trained on these hyperparameters and the pretrained gpt-2 was fine-tuned. The model was tested on a sample prompt relating to mental health and compared to the base pretrained gpt2 model.

The model was then further fine-tuned on a smaller dataset containing tweets and reddits for additional understanding of user prompts. Hyperparameter tuning was done again before this second tuning because the hyperparameters from the first tuning, did not quite fit this new data set. The model was also tested on a sample prompt.

#### 2.2.4 Statistical Evaluation and Analysis

Several scoring metrics were employed to assess the quality and performance of the text generation model on the dialogue validation dataset. These metrics serve the crucial role of determining the statistical significance of any observed differences or improvements in the model's responses, ensuring that the findings are not attributable to random chance.

The first two scoring metrics utilised were the BLEU Score and the ROUGE Score. The BLEU Score quantifies the similarity between the model's generated responses and human-generated responses and the ROUGE Score assesses the overlap and similarity between the model-generated text and reference texts. These metrics provide insights into how closely the model's outputs align with human-generated responses.

Another vital metric considered is Perplexity, which gauges the language model's ability to predict text. Perplexity was computed on the validation dataset by preprocessing and tokenizing the data, then employing the model to calculate the perplexity score, indicating the model's predictive capabilities.

Semantic Similarity was assessed using a BERT-based model by encoding both the human-generated and model-generated responses, obtaining numerical embeddings, and calculating the cosine similarity between these embeddings. This metric provides insights into the semantic similarity between the model's responses and human-generated references, shedding light on the model's ability to generate contextually relevant content.

Sentiment analysis was also performed to obtain the human and model sentiment subjectivity and polarity. To assess the statistical significance of the computer-generated responses, statistical tests were conducted. The Shapiro-Wilk tests and Wilcoxon Rank-Sum Test (Mann-Whitney U Test) was employed to find statistical significance between the groups, suggesting meaningful distinctions in sentiment scores between human and model-generated responses.

```python
def remove_empty_strings(df, col):
    # Find the count of empty strings and '[ Removed by Reddit ]' in the specified column
    empty_string_count = (df[col] == '').sum()
    removed_by_reddit_count = (df[col] == '[ Removed by Reddit ]').sum()

    if empty_string_count > 0 or removed_by_reddit_count > 0:
        # Remove rows with empty strings or '[ Removed by Reddit ]' in the specified column
        df.drop(df[(df[col] == '') | (df[col] == '[ Removed by Reddit ]')].index, inplace=True)
        print(f"Removed {empty_string_count} rows with empty strings and {removed_by_reddit_count} rows with '[ Removed by Reddit ]' from '{col}' column in-place.")
    else:
        print(f"No empty strings or '[ Removed by Reddit ]' found in '{col}' column.")
    return
```

```python
def remove_duplicates(df, col):
    # Find and count duplicate values in the specified column
    duplicate_counts = df[col].duplicated(keep='first')

   # Count the number of duplicates removed
    num_duplicates_removed = duplicate_counts.sum()

    # Remove rows with duplicate values in place
    df.drop(df[duplicate_counts].index, inplace=True)

    # Reset the index of the DataFrame (optional)
    df.reset_index(drop=True, inplace=True)

    print (num_duplicates_removed, 'duplicates removed')
```

```python
def remove_some_punct(text):
    # Remove the date at the beginning of the text (if applicable)
    text = re.sub(r'^\w+\s+\d+,\s+\d+\s+', '', text)

    # Remove <p> from the start of the string (if it exists)
    text = re.sub(r'^<p>', '', text)

    # Split the text into words
    words = text.split()

    # Filter out words starting with '#' but not numbers
    filtered_words = [word for word in words if not (word.startswith('#') and not word[1:].isdigit())]

    # Join the filtered words with spaces
    text = ' '.join(filtered_words)

    # Separate words joined by a full stop
    text = re.sub(r'\.(?=\S)', '. ', text)

    # Replace \xa0 with spaces
    text = text.replace('\xa0', ' ')

    # Define a regex pattern to remove all punctuation except for specified characters
    pattern = r"[^\w\s.,':\-()\n]"

    # Use regex to remove punctuation
    text = re.sub(pattern, '', text)

    return text
```

```python
class DialogueDataset(Dataset):
    def __init__(self, data, tokenizer, max_length=512):
        self.data = data
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        parts = item.split(" Response: ")

        # Check if the parts list contains at least two elements
        if len(parts) < 2:
            # Handle cases where the format is incorrect by returning an empty tensor
            return {
                "input_ids": torch.tensor([], dtype=torch.long),
            }

        question = parts[0].replace("Question: ", "")

        # Combine all remaining parts as the response
        response = " ".join(parts[1:])

        # Tokenize the input text
        input_text = f"Question: {question} Response: {response}"
        input_ids = self.tokenizer.encode(
            input_text,
            add_special_tokens=True,
            max_length=self.max_length,
            truncation=True,
            padding="max_length",
            return_tensors="pt"
        )

        return {
            "input_ids": input_ids.squeeze(),
        }

def dialogue_data_loader(texts, tokenizer, batch_size=16, max_length=512, shuffle=True):
    dialogue_dataset = DialogueDataset(formatted_texts, tokenizer, max_length)

    # Split the dataset into training and validation sets
    train_size = int(0.8 * len(dialogue_dataset))
    val_size = len(dialogue_dataset) - train_size
    dialogue_train_dataset, dialogue_val_dataset = random_split(dialogue_dataset, [train_size, val_size])

    # Prepare the datasets for training and validation on CUDA (if available)
    data_collator = DataCollatorForLanguageModeling(
        tokenizer=tokenizer, mlm=False
    )

    # Create DataLoaders for training and validation on CUDA (if available)
    dialogue_train_loader = DataLoader(
        dialogue_train_dataset,
        batch_size=batch_size,
        shuffle=shuffle,
        collate_fn=data_collator,
    )
    dialogue_val_loader = DataLoader(
        dialogue_val_dataset,
        batch_size=batch_size,
        shuffle=False,
        collate_fn=data_collator,
    )

    return dialogue_train_dataset, dialogue_val_dataset, dialogue_train_loader, dialogue_val_loader
```

```python
# Define a function for the objective (training) with hyperparameters
def objective(trial, train_dataset, val_dataset):
    # Define the hyperparameters to be tuned
    learning_rate = trial.suggest_loguniform("learning_rate", 1e-5, 1e-3)
    batch_size = trial.suggest_categorical("batch_size", [8, 16, 32])
    data_collator = DataCollatorForLanguageModeling(
        tokenizer=tokenizer, mlm=False)

    # Ensure that the batch size is an integer
    batch_size = int(batch_size)

    try:
        # Train and evaluate the model here

        # Define training arguments
        training_args = TrainingArguments(
            output_dir="my_gpt2",
            overwrite_output_dir=True,
            num_train_epochs=3,
            save_steps=10,  # Save model checkpoints every N steps
            evaluation_strategy="steps",
            eval_steps=10,  # Evaluate the model every N steps
            save_total_limit=2,
            learning_rate=learning_rate,
            report_to="tensorboard",  # Enable TensorBoard logging
        )

        # Create the model
        model = GPT2LMHeadModel.from_pretrained("gpt2")

        # Define the Trainer
        trainer = Trainer(
            model=model,
            args=training_args,
            data_collator=data_collator,
            train_dataset=train_dataset,
            eval_dataset=val_dataset,
        )

        # Train the model with early stopping
        trainer.train()

        # Retrieve the validation loss after training
        validation_loss = trainer.evaluate()["eval_loss"]

        return validation_loss

    except Exception as e:
        # Handle any exceptions that may occur during model training or evaluation
        # Return a high validation loss in case of an exception
        print(f"Error during trial: {str(e)}")
        return 1e6  # High validation loss to penalize this trial
```

```python
def train_gpt2_model(model_name, train_dataset, val_dataset, learning_rate, batch_size, num_train_epochs):
    # Load the pre-trained GPT-2 model and tokenizer
    model = GPT2LMHeadModel.from_pretrained(model_name)
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)
    tokenizer.pad_token = tokenizer.eos_token

    # Prepare the datasets for training and validation
    data_collator = DataCollatorForLanguageModeling(
        tokenizer=tokenizer, mlm=False
    )

    # Create TrainingArguments for the Trainer
    training_args = TrainingArguments(
        output_dir="my_gpt2",
        overwrite_output_dir=True,
        num_train_epochs=num_train_epochs,
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        learning_rate=learning_rate,
        save_steps=10,  # Save model checkpoints every N steps
        evaluation_strategy="steps",
        eval_steps=10,  # Evaluate the model every N steps
        save_total_limit=2,
        report_to="tensorboard",  # Enable TensorBoard logging
        gradient_accumulation_steps=4,
    )

    # Create the Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
    )

    # Start training
    trainer.train()

    # Save the trained model
    trainer.save_model()

    # Optionally, evaluate the model on a validation dataset
    trainer.evaluate()

    return trainer
```

```python
def processed_generated_response(prompt, model_path=None):
    if model_path:
        # Call generate response function
        generated_response = generate_response(model_path, prompt)
    else:
        generated_response = prompt

    # Extract text after 'Response:'
    response_text = re.sub(r'^.*?Response:', '', generated_response)

    # Split the text into sentences using a simple regex
    sentences = re.split(r'\s*[.!?]\s*', response_text)

    # Count the occurrences of each sentence
    sentence_counts = Counter(sentences)

    # Filter sentences that occur only once
    non_repeated_sentences = [sentence for sentence in sentences if sentence_counts[sentence] == 1]

    # Join the non-repeated sentences back into a single text
    result_text = ' '.join(non_repeated_sentences)

    return result_text

```

## 6 Statistical Evaluation and Analysis

To determine whether any observed differences or improvements are statistically significant or if they could have occurred by random chance, statistical tests were carried out on the dialogue validation dataset based on certain scoring metrics that are commonly used for evaluating text generation quality. These metrics are:
1. **BLEU Score**: Measures the similarity between a generated text and one or more reference texts. The human generated responses were used a reference and the model generated responses were used as the hypothesis to obtain the BLEU score.
2. **ROUGE Score**: Evaluates the overlap and similarity between the generated text and reference texts, often used for summarisation. The ROUGE score was also obtained using the human generated response and the computer generated text as the hypothesis.
3. **Perplexity**: Measures the language model's ability to predict the text; lower perplexity indicates better quality. The perplexity was calculated on the validation dataset by loading the model, tokenizing the validation data, processesing it in batches, and computing the perplexity score as a measure of how well the model predicts the text in the validation dataset.
4. **Semantic Similarity**: Compute the semantic similarity between the generated text and reference text using pre-trained models like BERT or Word2Vec. The semantic similarity score was calculated using a pre-trained BERT-based model (default: "paraphrase-MiniLM-L6-v2"). The function encodes two input sentences which were the human and model responses, using the model to obtain numerical embeddings and calculates the cosine similarity between these embeddings.

In order to calculate these metrics, model responses were generated on all the questions from the dialogue validation dataset. These respones were saved into a dataframe along with the questions and human generated responses. The Bleu, rogue and semantic similarity scores were calculated for each row of the data frame. The final scores were a mean of all the rows in each metric's column. The scores obtained were different from expectations.

The bleu and rogue scores were far lower than expected. Typically, higher BLEU scores are better, but the exact threshold for a "good" BLEU score can vary depending on the specific NLP task. Low BLEU and ROUGE scores can occur despite a good model. This can be primarily attributed to the diversity of responses. If a model can generate various valid answers for a given input, but the reference data offers only a limited set of responses, these metrics may penalise the model for not matching the references precisely. Additionally, issues related to reference quality, task complexity, ambiguity, and fluency can also contribute to lower scores.

The perplexity score was very low which indicates that the model is making reasonably accurate predictions, with little level of uncertainty or variability in its output. Although this was performed on a validation set, such a score might point to some overfitting of the model.

The semantic similarity score was average, indicating a moderate level of agreement between human and computer-generated responses. While it suggests some commonality in meaning and content, there is also notable divergence. Ultimately, there is room for improvement in aligning the two types of responses for better consistency and relevance.

Sentiment analysis was also performed to obtain the human and model sentiment subjectivity and polarity. Similar model and human sentiment subjectivity values were obtained which implies that both humans and the tuned model are expressing similar emotions (sentiment) and are similarly subjective or objective in their text. Markedly different values for human and model sentiment polarity were obtained which indicates a disagreement or mismatch between the sentiment (emotional tone) expressed by human evaluators and the sentiment predicted or generated by a machine learning mode. This can be due to ambiguity in text, model limitations, dataset bias and subjectivity

Given these metrics statistical tests were performed to determine if the computer generated responses were statistically significant. Shapiro Wilk tests indicated non-normality in the computed metrics. Hence, to find statistical difference between model and human sentiment scores a  Wilcoxon Rank-Sum Test (Mann-Whitney U Test) was used instead of a paired T-test since paired T-tests assume differences between paired observations follow a normal distribution. the Wilcoxon Rank-Sum Test indicated statistical significance between groups. Some relevant visualisations were also plotted.

```python
# Define a function to calculate BLEU and ROUGE-L scores
def calculate_scores(reference, hypothesis):
    '''Function to calculate Bleu and Rouge scores of the model'''

    # BLEU Score
    bleu_score = sentence_bleu([reference], hypothesis)

    # ROUGE-L Score
    scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
    scores = scorer.score(reference, hypothesis)
    rouge_l_score = scores['rougeL'].fmeasure

    return bleu_score, rouge_l_score
```

```python
# Define the function to calculate semantic similarity
def calculate_semantic_similarity(sentence1, sentence2, model_name="paraphrase-MiniLM-L6-v2"):
    """
    Calculate semantic similarity between two sentences using a pre-trained BERT-based model.

    Args:
        sentence1 (str): The first sentence.
        sentence2 (str): The second sentence.
        model_name (str): The name of the pre-trained model to use.

    Returns:
        float: The semantic similarity score (cosine similarity).
    """
    # Load a pre-trained BERT model for semantic similarity
    embedder = SentenceTransformer(model_name)

    # Encode the sentences to obtain their embeddings
    embeddings1 = embedder.encode(sentence1, convert_to_tensor=True)
    embeddings2 = embedder.encode(sentence2, convert_to_tensor=True)

    # Calculate cosine similarity between the embeddings
    cosine_score = util.pytorch_cos_sim(embeddings1, embeddings2).item()

    return cosine_score
```

### 6.4 Statistical Tests

The previously calculated metrics can be used to perform statistical tests on the data. These statistical tests include:<br>

1. Shapiro-Wilk Test: This is a test for normality. When data is normally distributed, it benefits statistical tests, simplifies calculations, and makes inferences more reliable. The data obtained from all the computed metrics were not normally distributed so a Wilcoxon rank-sum was used to find statistical significance between groups.
2. Wilcoxon rank-sum test: This test was used to find statistical significance between the human and model sentiment polarity and subjectivity. The test concludes that there is a statistically significant difference between the groups.

#### 6.4.1 Shapiro-Wilk Test

```python
def check_normality(df, column_name, alpha=0.05):
    """
    Check the normality of a DataFrame column using the Shapiro-Wilk test.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the data.
    column_name (str): The name of the column to test for normality.
    alpha (float): The significance level for the test (default is 0.05).

    Returns:
    str: A message indicating whether the data appears to be normally distributed.
    """

    data = df[column_name]
    statistic, p_value = stats.shapiro(data)

    if p_value > alpha:
        return f"The data in column '{column_name}' appears to be normally distributed (p-value > {alpha})"
    else:
        return f"The data in column '{column_name}' does not appear to be normally distributed (p-value <= {alpha})"
```

```python
# Check normality for the numeric columns in the dataframe
for col in df_gen_res.columns[3:]:
  print(check_normality(df_gen_res, col))
```

#### 6.4.2  Wilcoxon Rank-Sum Test (Mann-Whitney U Test)

```python
def wilcoxon_rank_sum_test(group1, group2, alpha=0.05):
    """
    Perform the Wilcoxon Rank-Sum Test (Mann-Whitney U Test) between two groups.

    Parameters:
        group1 (array-like): First group of data.
        group2 (array-like): Second group of data.
        alpha (float): Significance level (default is 0.05).

    Returns:
        dict: A dictionary containing the test results.
            - 'Test Statistic': The test statistic.
            - 'P-Value': The p-value.
            - 'Conclusion': A string indicating whether to reject or fail to reject the null hypothesis.

    Example:
        test_results = wilcoxon_rank_sum_test(data_group1, data_group2)
    """
    # Perform the Wilcoxon Rank-Sum Test
    statistic, p_value = stats.ranksums(group1, group2)

    # Check if the p-value is significant
    if p_value < alpha:
        conclusion = "Reject the null hypothesis: There is a significant difference between the groups."
    else:
        conclusion = "Fail to reject the null hypothesis: There is no significant difference between the groups."

    # Prepare and return the test results as a dictionary
    test_results = {
        'Test Statistic': statistic,
        'P-Value': p_value,
        'Conclusion': conclusion
    }

    return test_results
```
